```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import datetime as dt
```

```python
covid = pd.read_csv('covid.csv', parse_dates=['Date'], dayfirst=True)
```

```python
covid.head()
```

```python
covid.rename(columns={"Date": "date", "State/UnionTerritory": "state", "Cured": "cured","Deaths":"deaths","Confirmed":"confirmed"},inplace=True,)
covid
```

```python
covid.head()
```

```python
covid.tail()
```

```python
today = covid[covid.date=='2021-08-11']
```

```python
today
```

```python
max_confirmed_case = today.sort_values(by='confirmed',ascending=False)
```

```python
max_confirmed_case
```

```python
top_states_confirmed=max_confirmed_case[0:5]
```

```python
top_states_confirmed
```

```python
# making bar-plot for states with top confirmed cases
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x='state',y='confirmed',data=top_states_confirmed, hue='state')
plt.show()
```

```python
#sorting the data w.r.t no. of deaths cases
max_death_cases = today.sort_values(by='deaths',ascending=False)
max_death_cases
```

```python
#getting states with maximum no. of deaths cases
top_states_death=max_death_cases[0:5]
```

```python
# making bar-plot for states with top deaths cases 
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x='state',y='deaths',data=top_states_death,hue='state')
plt.show()
```

```python
#sorting data w.r.t no. of cured cases
max_cured_cases=today.sort_values(by='cured',ascending=False)
max_cured_cases
```

```python
#getting states with maximum no. of cured cases
top_states_cured=max_cured_cases[0:5]
```

```python
#making bar-plot for states with top death cases 
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='state',y='cured',data=top_states_cured,hue='state')
plt.show()
```
